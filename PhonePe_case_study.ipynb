# 1.3 check missing values

```python
# checking missing values in each dataset
state_transaction.isnull().sum()
state_transaction_types.isnull().sum()
state_devicedata.isnull().sum()
District_transaction.isnull().sum()
District_Demographics.isnull().sum()
```

```python
# calculating the column with missing value has outliers or not
Q1 = state_transaction['Amount (INR)'].quantile(0.25)  # First quartile (25th percentile)
Q3 = state_transaction['Amount (INR)'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile range
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
outliers = state_transaction[(state_transaction['Amount (INR)'] < lower_bound) | (state_transaction['Amount (INR)'] > upper_bound)]
outliers
```

```python
# calculate median to replace missing values
median_value=state_transaction['Amount (INR)'].sort_values().median()
median_value
state_transaction['Amount (INR)']=state_transaction['Amount (INR)'].fillna(median_value)
state_transaction.isnull().sum()
```

```python
# demonstrate code and ATV column to replace missing values
# District_transaction[District_transaction['Code'].isnull()]
District_transaction['Code']=District_transaction['Code'].fillna('unknown')
District_transaction
District_transaction.isnull().sum()
```

```python
# Handle ATV (INR) column missing values
# District_transaction[District_transaction['ATV (INR)'].isnull()]
District_transaction['ATV (INR)']=District_transaction['ATV (INR)'].fillna(0)
District_transaction.isnull().sum()
```

```python
# 2 Calculate the percentage of missing values for each column that has missing values.
Percentage=1/state_transaction['Amount (INR)'].count()*100
print(f"Percentage of missing values in amount column of State_Txn and Users dataset is {Percentage.round(3)}%")
Percentage=28/District_transaction['Code'].count()*100 
print(f"Percentage of missing values in code column of District_Txn and Users dataset is {Percentage.round(3)}%")
Percentage=4/District_transaction['ATV (INR)'].count()*100
print(f"Percentage of missing values in ATV (INR) column of District_Txn and Users dataset is {Percentage.round(3)}%")
```

```python
# Highlight which column has the highest percentage of missing values in each dataset
state_txn_missing = {
    "Amount (INR)": (1 / state_transaction['Amount (INR)'].count() * 100)
}
district_txn_missing = {
    "Code": (28 / District_transaction['Code'].count() * 100),
    "ATV (INR)": (4 / District_transaction['ATV (INR)'].count() * 100)
}
state_txn_missing_series = pd.Series(state_txn_missing)
district_txn_missing_series = pd.Series(district_txn_missing)

# Find the column with the maximum missing percentage using idxmax()
state_txn_max_missing = state_txn_missing_series.idxmax()
district_txn_max_missing = district_txn_missing_series.idxmax()
print(f"Column with the highest missing percentage in State_Txn and Users : {state_txn_max_missing}")
print(f"Column with the highest missing percentage in District_Txn and Users : {district_txn_max_missing}")
```

# Task 4: Data Merging and Advaned anlysis

##  4.1 Ratio of users of population by  state.

```python
# Merge
state_users=state_transaction.groupby('State')['Registered Users'].sum().reset_index()
state_population=District_Demographics.groupby('State')['Population'].sum().reset_index()
merge_df= pd.merge(state_users,state_population , on='State', how='inner')
```

```python
plt.figure(figsize = (16,9))
merge_df['ratio']=(merge_df['Registered Users']/merge_df['Population'])
merge_df
sorted_df=merge_df.sort_values('ratio').nlargest(5,'ratio')
plt.bar(sorted_df['State'],sorted_df['ratio'],color='green')
plt.title("Top 5 states with ratio of registerd users to the population")
plt.xlabel("State")
plt.ylabel("Transactions") 
plt.tight_layout()
plt.show()
```

## 4.2 Correlate population density with transaction volume

```python
# 1.  Merge two datasets
distinct_users=District_transaction.groupby(['State','District'])['Transactions'].sum().reset_index()
distinct_population=District_Demographics.groupby(['State','District'])['Density'].sum().reset_index()
df= distinct_population.merge(distinct_users, on=['State','District'], how='left').fillna(0)
df
```

```python
# 2: Calculate the correlation b/w population density and transaction volume.
df.columns
df['Density'].corr(df['Transactions']).round(2)
```

```python
#3: Create a scater plot to visualize the correlation b/w population density and transaction volume.
sns.scatterplot(x="Density",
                y="Transactions",
                data=df)
```

## 4.3: Average transaction amount per user. 


```python
#1.Calculate average transaction amount per user for each state.
state_users=state_transaction.groupby('State')[['Amount (INR)','Registered Users']].sum().reset_index()
state_users['Average_Transaction_amount']=(state_users['Amount (INR)']/state_users['Registered Users']).round(2)
state_users.loc[:,['Average_Transaction_amount']]
```

```python
# 2. Identify the top 5 states with highest average transaction amount per user and the lowest transaction amount per user.
top_states=state_users.nlargest(5,'Average_Transaction_amount')
top_states[['State']].reset_index(drop=True)
# lowest 5 states
lower_states=state_users.nsmallest(5,'Average_Transaction_amount')
lower_states[['State']].reset_index(drop=True)
```

## 4.4 Device brand usage ratio

```python
# 1. merge datasets
level_1=state_transaction.groupby('State')['Registered Users'].sum().reset_index()
level_2=state_devicedata.groupby(['State','Brand']).agg(No_of_users=('Registered Users','sum')).reset_index() 
df= level_2.merge(level_1, on=['State'],how='inner')    
df
```

```python
# 2. Calculate the ratio of users of each device brand to the total no. of registered users in each state.
df['ratio']=df['No_of_users']/df['Registered Users']
df[['ratio']]

#  3.Create the bar plot depiciting the device brand usage ratio for each state.
plt.figure(figsize=(16,9))
top_states = level_1.nlargest(5, 'Registered Users')['State']

# Step 2: Filter `df` for the top 5 states
filtered_df = df[df['State'].isin(top_states)]
filtered_df

# Step 3: Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_df, x='State', y='ratio', hue='Brand', palette='viridis')

plt.title('Device Brand Usage Ratio by State (Top 5 States)', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Usage Ratio', fontsize=14)
plt.xticks(rotation=45)
plt.show()
# pip install scipy
# pip install scikit-learn
```

# Task 5: Data Visualization

## 5.1 Plot the total no. of  transactions and amount over time for a selected state.

```python
# Create a line chart showing the  total no. of  transactions and amount over time(years and quarters) for a selected state
# select state which has max no.of transaction
max_transaction_state = (
    state_transaction.groupby('State')['Transactions']
    .sum()
    .idxmax()
)
max_transaction_state
# filter the data
filtered_data = state_transaction[state_transaction['State'] == max_transaction_state]
df_agg=filtered_data.groupby(['Year','Quarter']).agg(Total_Transactions=('Transactions','sum'),
                                       Total_amount=('Amount (INR)','sum')).reset_index()
df_agg['Time']=df_agg['Year'].astype(str) + " Q" + df_agg['Quarter'].astype(str)
df_agg
# plotting the data
plt.figure(figsize=(16,9))
plt.plot(df_agg['Time'],df_agg['Total_Transactions'])
plt.title("Showing total no. of transactions and amount over time(years and quarters) for selected state")
plt.xlabel("Time(over years and quaters)", fontsize=14)
plt.ylabel('Total Transaction amount', fontsize=14)
plt.show()
```
